## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
df2009=pd.read_csv('DOM2009.csv',sep=',')
df2013=pd.read_csv('DOM2013.csv',sep=',')
df2015=pd.read_csv('DOM2015.csv',sep=',')
```

```python
df2015["Regiao"] = pd.Series(len(df2015))
df2015["Classe_Social"] = pd.Series(len(df2015))

df2013["Regiao"] = pd.Series(len(df2013))
df2013["Classe_Social"] = pd.Series(len(df2013))

df2009["Regiao"] = pd.Series(len(df2009))
df2009["Classe_Social"] = pd.Series(len(df2009))
```

```python
#2015
df2015["REGIAO"]='x'
df2015.loc[(df2015['UF'] >=10) & (df2015['UF'] < 20),'REGIAO'] = 'Norte'
df2015.loc[(df2015['UF'] >=20) & (df2015['UF'] < 30),'REGIAO'] = 'Nordeste'
df2015.loc[(df2015['UF'] >=30) & (df2015['UF'] < 40),'REGIAO'] = 'Sudeste'
df2015.loc[(df2015['UF'] >=40) & (df2015['UF'] < 50),'REGIAO'] = 'Sul'
df2015.loc[(df2015['UF'] >=50) & (df2015['UF'] < 60),'REGIAO'] = 'Centro-Oeste'

df2015.loc[(df2015.V4614 > 45920),'Classe_Social'] = "Classe A1"
df2015.loc[(df2015.V4614 > 33180) & (df2015.V4614 <= 45920),'Classe_Social'] = "Classe A2"
df2015.loc[(df2015.V4614 > 19016) & (df2015.V4614 <= 33180),'Classe_Social'] = "Classe B1"
df2015.loc[(df2015.V4614 > 10624) & (df2015.V4614 <= 19016),'Classe_Social'] = "Classe B2"
df2015.loc[(df2015.V4614 > 5836) & (df2015.V4614 <= 10624),'Classe_Social'] = "Classe C1"
df2015.loc[(df2015.V4614 > 3848) & (df2015.V4614 <= 5836),'Classe_Social'] = "Classe C2"
df2015.loc[(df2015.V4614 > 2720) & (df2015.V4614 <= 3848),'Classe_Social'] = "Classe D"
df2015.loc[(df2015.V4614 > 1660) & (df2015.V4614 <= 2720),'Classe_Social'] = "Classe E"

#2013
df2013["REGIAO"]='x'
df2013.loc[(df2013['UF'] >=10) & (df2013['UF'] < 20),'REGIAO'] = 'Norte'
df2013.loc[(df2013['UF'] >=20) & (df2013['UF'] < 30),'REGIAO'] = 'Nordeste'
df2013.loc[(df2013['UF'] >=30) & (df2013['UF'] < 40),'REGIAO'] = 'Sudeste'
df2013.loc[(df2013['UF'] >=40) & (df2013['UF'] < 50),'REGIAO'] = 'Sul'
df2013.loc[(df2013['UF'] >=50) & (df2013['UF'] < 60),'REGIAO'] = 'Centro-Oeste'

df2013.loc[(df2013.V4614 > 45920),'Classe_Social'] = "Classe A1"
df2013.loc[(df2013.V4614 > 33180) & (df2013.V4614 <= 45920),'Classe_Social'] = "Classe A2"
df2013.loc[(df2013.V4614 > 19016) & (df2013.V4614 <= 33180),'Classe_Social'] = "Classe B1"
df2013.loc[(df2013.V4614 > 10624) & (df2013.V4614 <= 19016),'Classe_Social'] = "Classe B2"
df2013.loc[(df2013.V4614 > 5836) & (df2013.V4614 <= 10624),'Classe_Social'] = "Classe C1"
df2013.loc[(df2013.V4614 > 3848) & (df2013.V4614 <= 5836),'Classe_Social'] = "Classe C2"
df2013.loc[(df2013.V4614 > 2720) & (df2013.V4614 <= 3848),'Classe_Social'] = "Classe D"
df2013.loc[(df2013.V4614 > 1660) & (df2013.V4614 <= 2720),'Classe_Social'] = "Classe E"


#2009
df2009["REGIAO"]='x'
df2009.loc[(df2009['UF'] >=10) & (df2009['UF'] < 20),'REGIAO'] = 'Norte'
df2009.loc[(df2009['UF'] >=20) & (df2009['UF'] < 30),'REGIAO'] = 'Nordeste'
df2009.loc[(df2009['UF'] >=30) & (df2009['UF'] < 40),'REGIAO'] = 'Sudeste'
df2009.loc[(df2009['UF'] >=40) & (df2009['UF'] < 50),'REGIAO'] = 'Sul'
df2009.loc[(df2009['UF'] >=50) & (df2009['UF'] < 60),'REGIAO'] = 'Centro-Oeste'

df2009.loc[(df2009.V4614 > 45920),'Classe_Social'] = "Classe A1"
df2009.loc[(df2009.V4614 > 33180) & (df2009.V4614 <= 45920),'Classe_Social'] = "Classe A2"
df2009.loc[(df2009.V4614 > 19016) & (df2009.V4614 <= 33180),'Classe_Social'] = "Classe B1"
df2009.loc[(df2009.V4614 > 10624) & (df2009.V4614 <= 19016),'Classe_Social'] = "Classe B2"
df2009.loc[(df2009.V4614 > 5836) & (df2009.V4614 <= 10624),'Classe_Social'] = "Classe C1"
df2009.loc[(df2009.V4614 > 3848) & (df2009.V4614 <= 5836),'Classe_Social'] = "Classe C2"
df2009.loc[(df2009.V4614 > 2720) & (df2009.V4614 <= 3848),'Classe_Social'] = "Classe D"
df2009.loc[(df2009.V4614 > 1660) & (df2009.V4614 <= 2720),'Classe_Social'] = "Classe E"
```

```python
#Somente Casas em área Urbana
Casas2015 = df2015[(df2015.V0202 == 2) & (df2015.V4105 == 1)]
Casas2013 = df2013[(df2013.V0202 == 2) & (df2013.V4105 == 1)]
Casas2009 = df2009[(df2009.V0202 == 2) & (df2009.V4105 == 1)]
```

```python
Casas2015.head()
```

```python
classes2015 = Casas2015.Classe_Social.value_counts(True)*100
classes2015.plot(kind="pie")
print(classes2015)
plt.title('Organização Social No Brasil em 2015 em %')
```

```python
classes2013 = Casas2013.Classe_Social.value_counts(True)*100
classes2013.plot(kind="pie")
print(classes2013)
plt.title('Organização Social No Brasil em 2013 em %')
```

```python
classes2009 = Casas2009.Classe_Social.value_counts(True)*100
classes2009.plot(kind="pie")
print(classes2009)
plt.title('Organização Social No Brasil em 2009 em %')
```

## Número de cômodos da casa

```python
comodos2015 = pd.crosstab(Casas2015.Classe_Social,Casas2015.V0204)
comodos2015.plot(kind="bar")
plt.xlabel("Classe Social")
plt.ylabel("Números de Domicílios com x numero de comodos")
plt.title('Número de Cômodos por classe Social 2015')
plt.show()
```

```python
comodos2013 = pd.crosstab(Casas2013.Classe_Social,Casas2013.V0204)
comodos2013.plot(kind="bar")
plt.xlabel("Classe Social")
plt.ylabel("Números de Domicílios com x numero de comodos")
plt.title('Número de Cômodos por classe Social 2015')
plt.show()
```

```python
comodos2009 = pd.crosstab(Casas2009.Classe_Social,Casas2009.V0204)
comodos2009.plot(kind="bar")
plt.xlabel("Classe Social")
plt.ylabel("Números de Domicílios com x numero de comodos")
plt.title('Número de Cômodos por classe Social 2015')
plt.show()
```

```python
print('2015')
print(comodos2015)
print()
print('2013')
print(comodos2013)
print()
print('2009')
print(comodos2009)
```

## Número de Banheiros

```python
numero_banheiro2015 = pd.crosstab(Casas2015.Classe_Social,Casas2015.V2016)
numero_banheiro2015.plot(kind='bar')
plt.title('Numero de Banheiro por domicílio 2015')
print(numero_banheiro2015)
```

```python
numero_banheiro2013 = pd.crosstab(Casas2013.Classe_Social,Casas2013.V2016)
numero_banheiro2013.plot(kind='bar')
plt.title('Numero de Banheiro por domicílio 2013')
print(numero_banheiro2013)
```

```python
numero_banheiro2009 = pd.crosstab(Casas2009.Classe_Social,Casas2009.V2016)
numero_banheiro2009.plot(kind='bar')
plt.title('Numero de Banheiro por domicílio 2009')
print(numero_banheiro2009)
```
