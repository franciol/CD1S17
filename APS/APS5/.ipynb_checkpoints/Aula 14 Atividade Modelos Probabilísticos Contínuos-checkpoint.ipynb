```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import scipy as sp
from numpy import mean, var, arange
import numpy as np
from matplotlib import pyplot


print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
PopBairros = pd.read_excel('Padaria.xlsx')
d0 = pd.DataFrame()
d0['A'] = sorted(PopBairros.A)
d0['B'] = sorted(PopBairros.B)
```

```python
plt.plot(d0.A, 'b')
plt.plot(d0.B, 'r')
plt.legend()
plt.title('Grafico Renda crescente dos Bairros')
plt.xlabel('Individuo')
plt.ylabel('Renda')
```

```python
d0.describe()
```

```python
d0.std()
```

### Bairro B:
maior Variação de renda no bairro, tendo os mais pobres e os mais ricos dos conjuntos de amostra. 

```python
faixas =np.arange(9,19,0.5)
```

```python

d0.plot.hist(bins=faixas,title='Histograma Distribuição de Renda nos Bairros', figsize=(6, 6), alpha=0.5, normed=1, color=(('y'),('black')))
plt.xlabel('Renda (em 1000 reais)')
plt.legend(('Bairro B','Bairro A'), loc='best')
```

# 4) 
## A)

```python
d0.A.hist(bins = faixas, color = "b", alpha = 0.6, normed = 1)
x = sorted(stats.norm.rvs(size = 10000, loc = 15.32, scale = 0.59))
plt.plot(x, stats.norm.pdf(x, loc = 15.32, scale = 0.60), lw = 3, alpha = 0.6)
```

```python
d0.B.hist(bins = faixas, color = "r", alpha = 0.6, normed = 1)
y = sorted(stats.norm.rvs(size = 10000, loc = 14.22, scale = 1.59))
plt.plot(y, stats.norm.pdf(y, loc = 14.22, scale = 1.84), lw = 3, alpha = 0.6)
```

# B)

```python
freqrelat = []
for i in range (1, 501):
    freqrelat.append ((i-0.5)/500)
```

```python
freqnormalA = []
for i in d0.A:
    freqnormalA.append(stats.norm.cdf (i, loc = 15.32, scale = 0.59))
```

```python
plt.figure(figsize=(10, 10))
plt.plot (d0.A, freqrelat, lw = 1.5, color = 'r', alpha = 0.7, label = 'Frequência empírica')
plt.plot (d0.A, freqnormalA, lw = 1.5, color = 'b', alpha = 0.7, label = 'Frequência na distribuição normal')
pyplot.legend (loc = 'best')
plt.ylabel ('Frequência relativa (quantil)')
plt.xlabel ('Valores das rendas')
plt.title ('Região A')
```

```python
freqnormalB = []
for i in d0.B:
    freqnormalB.append(stats.norm.cdf (i, loc = 14.22, scale = 1.89))
```

```python
plt.plot (d0.B, freqrelat, lw = 1.5, color = 'r', alpha = 0.7, label = 'Frequência empírica')
plt.plot (d0.B, freqnormalB, lw = 1.5, color = 'b', alpha = 0.7, label = 'Frequência na distribuição normal')
pyplot.legend (loc = 'best')
plt.ylabel ('Frequência relativa (quantil)')
plt.xlabel ('Valores das rendas')
plt.title ('Região B')
```

Observa-se nos gráficos plotados que o bairro A tem uma curva mais parecida com uma normal, enquanto o bairro B é menos consistente com essa distribuição. Assim A pode ser considerada uma distribuição Normal e B tambem poderia, levando-se em conta uma porceentagem de erro.

# 5)
## A)

```python
print ('Probabilidade de uma família da região A ter uma renda maior que 17 mil reais:', 1- (stats.norm.cdf (17, loc = 15.32, scale = 0.59)))
print ('Probabilidade de uma família da região A ter uma renda maior que 17 mil reais:', 1- (stats.norm.cdf (17, loc = 14.22, scale = 1.84)))
```

## B)

```python
print ('Probabilidade de uma família da região A ter uma renda maior que 16 mil reais dado que ela possui uma renda > 15 mil reais:', (((stats.norm.cdf (16, loc = 15.32, scale = 0.59) - (stats.norm.cdf (15, loc = 15.32, scale = 0.59)))/((stats.norm.cdf (18, loc = 15.32, scale = 0.59))-(stats.norm.cdf (15, loc = 15.32, scale = 0.59))))))
print ('Probabilidade de uma família da região B ter uma renda maior que 16 mil reais dado que ela possui uma renda > 15 mil reais:', (((stats.norm.cdf (16, loc = 14.22, scale = 1.84) - (stats.norm.cdf (15, loc = 14.22, scale = 1.84)))/((stats.norm.cdf (18, loc = 14.22, scale = 1.84))-(stats.norm.cdf (15, loc = 14.22, scale = 1.84))))))
```

## C)

```python
print ('Renda da família com menor renda dentro dos 5% com maior renda da região A:', stats.norm.ppf (0.95, loc = 15.32, scale = 0.59))
```

```python
print ('Renda da família com menor renda dentro dos 5% com maior renda da região B:', stats.norm.ppf (0.95, loc = 14.22, scale = 1.84))
```

```python
freqrelat.index(0.949)
print ('Obtidos apenas pelo índex da lista, sem utilizar a normal')
print ('Renda da família com menor renda dentro dos 5% com maior renda da região A:', d0.A[475])
print ('Renda da família com menor renda dentro dos 5% com maior renda da região B:', d0.B[475])
```

Comparando os valores dos 5% mais pobres do bairro A, a distribuição se apresenta muito parecida com a normal, já a B tem uma varição maior, que demonstra que ela não é uma distribuição Normal, somente parecida.
