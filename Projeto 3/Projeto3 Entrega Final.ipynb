## Estudo da Distribuição da Média Amostral  $\bar{X}$  via SIMULAÇÃO

## 1)
### Usar Poisson

```python
poisson = stats.poisson.rvs(mu=5 , loc=0,size=10000)
faixas = range(0,20,1)

plt.hist(poisson,normed=True,bins=faixas,alpha=0.6,color='#DAAC50')
```

```python
var = poisson.var()
DP = var**(1/2)
media_entrada = poisson.mean()
print('Variância entrada de clientes: ',var)
print('Desvio Padrão entrada clientes: ',DP)
print('Média de entrada: ',media_entrada)
```

### 2)

```python
n = 2
lista_media = []
media_lista = 0
var_lista = 0
for i in range(10000):
    a = np.random.choice(poisson,n)
    lista_media.append(a.mean())


media_das_medias = np.mean(lista_media)
var_medias = np.var(lista_media)
plt.hist(lista_media,color='#61369D',normed=True)
teta50 = var_medias**(1/2)
plt.hist(lista_media,color='#61369D',normed=True)
x = sorted(stats.norm.rvs(size = 10000, loc = media_das_medias, scale = teta50))
plt.plot(x, stats.norm.pdf(x, loc = media_das_medias, scale = teta50), lw = 3, alpha = 0.6)
plt.title('Distribuição média amostral para n=2')
plt.show()
print('Media das médias de n=2: ',media_das_medias)
print('Variancia das Medias de n=2: ',var_medias)

n = 4
lista_media = []
media_lista = 0
var_lista = 0
for i in range(10000):
    a = np.random.choice(poisson,n)
    lista_media.append(a.mean())


media_das_medias = np.mean(lista_media)
var_medias = np.var(lista_media)
teta50 = var_medias**(1/2)
plt.hist(lista_media,color='#61369D',normed=True)
x = sorted(stats.norm.rvs(size = 10000, loc = media_das_medias, scale = teta50))
plt.plot(x, stats.norm.pdf(x, loc = media_das_medias, scale = teta50), lw = 3, alpha = 0.6)
plt.hist(lista_media,color='#61369D',normed=True)
plt.title('Distribuição média amostral para n=4')
plt.show()
print('Media das médias de n=4: ',media_das_medias)
print('Variancia das Medias de n=4: ',var_medias)


n = 20
lista_media = []
media_lista = 0
var_lista = 0
for i in range(10000):
    a = np.random.choice(poisson,n)
    lista_media.append(a.mean())


media_das_medias = np.mean(lista_media)
var_medias = np.var(lista_media)
teta50 = var_medias**(1/2)
plt.hist(lista_media,color='#61369D',normed=True)
x = sorted(stats.norm.rvs(size = 10000, loc = media_das_medias, scale = teta50))
plt.plot(x, stats.norm.pdf(x, loc = media_das_medias, scale = teta50), lw = 3, alpha = 0.6)
plt.hist(lista_media,color='#61369D',normed=True)
plt.title('Distribuição média amostral para n=20')
plt.show()
print('Media das médias de n=20: ',media_das_medias)
print('Variancia das Medias de n=20: ',var_medias)


n = 50
lista_3_media = []
lista_15_media = []
lista_media = []
media_lista = 0
var_lista = 0
for i in range(10000):
    a = np.random.choice(poisson,n)
    lista_media.append(a.mean())
    lista_3_media.append(a[3])
    lista_15_media.append(a[15])

media_das_medias = np.mean(lista_media)
var_medias = np.var(lista_media)
teta50 = var_medias**(1/2)
plt.hist(lista_media,color='#61369D',normed=True)
x = sorted(stats.norm.rvs(size = 10000, loc = media_das_medias, scale = teta50))
plt.plot(x, stats.norm.pdf(x, loc = media_das_medias, scale = teta50), lw = 3, alpha = 0.6)
plt.title('Distribuição média amostral para n=50')
plt.show()
print('Media das médias de n=50: ',media_das_medias)
print('Variancia das Medias de n=50: ',var_medias)
```
