### Pergunta: 
    - Como espectativa de vida e renda per capita influenciam no IDH?

O objetivo desse trabalho é definir como as variações da espectativa de vida e renda per capita alteram o IDH de cada país do database, em 2011. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import statsmodels.api as sm1 
import statsmodels.formula.api as sm
%matplotlib inline
```

```python
IDH = pd.read_excel('Indicator_HDI.xlsx',index_col=0)
dadosEspecLife = pd.read_excel('indicator life_expectancy_at_birth.xlsx',index_col=0)
dadosRenda = pd.read_excel('indicator uncertainty_gdp_pc.xlsx',index_col=0)
```

```python
Renda = dadosRenda["A2011"]
Espectativa = dadosEspecLife["S2011"]
idh = IDH["I2011"]
```

```python
result = pd.concat([Renda,Espectativa,idh],axis=1,join="inner")
final = result.dropna()
```

```python
IDH = final.I2011
Espectativa = final.S2011
Renda = final.A2011
```

```python
# IDH
IDH.describe()
```

```python
# Espectativa de vida
Espectativa.describe()
```

```python
# Renda p/ Capita
Renda.describe()
```

```python
plt.figure(figsize=(20,10))


plt.subplot(121)
plt.plot(Renda,Espectativa,"o")
plt.title("QQ-plot Renda x Espectativa de Vida")
plt.xlabel("Renda per Capita")
plt.ylabel("Espectativa de Vida")

plt.subplot(122)
plt.plot(np.log(Renda),np.log(Espectativa),"o")
plt.title("QQ-plot Renda x Espectativa de Vida(Log)")
plt.xlabel("Renda per Capita")
plt.ylabel("Espectativa de Vida")

plt.plot()
print("Correlação: ",np.corrcoef(Renda,Espectativa)[0][1])
```

```python
plt.figure(figsize=(20,10))

plt.subplot(121)

plt.plot(Renda,IDH,"o")
plt.title("QQ-plot Renda x IDH")
plt.xlabel("Renda per Capita")
plt.ylabel("Indice de Desenvolvimento Humano(IDH)")


plt.subplot(122)
plt.plot(np.log(Renda),np.log(IDH),"o")
plt.title("QQ-plot Renda x IDH - Log")
plt.xlabel("Renda per Capita")
plt.ylabel("Indice de Desenvolvimento Humano(IDH)")

plt.plot()

print("Correlação: ",np.corrcoef(Renda,IDH)[0][1])
```

```python
plt.figure(figsize=(20,10))

plt.subplot(121)
plt.plot(Espectativa,IDH,"o")
plt.title("QQ-plot IDH x Espectativa de Vida")
plt.xlabel("Espectativa de Vida")
plt.ylabel("IDH")


plt.subplot(122)
plt.plot(np.log(Espectativa),np.log(IDH),"o")
plt.title("QQ-plot IDH x Espectativa de Vida - Log")
plt.xlabel("Espectativa de Vida")
plt.ylabel("IDH")

plt.plot()

print("Correlação: ",np.corrcoef(IDH,Espectativa)[0][1])
```

Como pode-se perceber, existe correlação entre os 3 datasets escolhidos. Aqueles com maior correlação no caso são espectativa e IDH, como pode se ver nos gráficos e no coeficiende de correlação. Como os gráficos Logarítmicos demonstraram melhor a correlação entre os modelos, será utilizado a escala logarítmica para o resto do projeto.

```python
Renda = np.log(Renda)
#IDH = np.log(IDH)
Espectativa = np.log(Espectativa)
final["LogRenda"] = Renda
final["LogIDH"] = IDH
final["LogEspect"] = Espectativa
```

```python
X = final[['LogRenda', 'LogEspect']]
y = final['LogIDH']

X = sm1.add_constant(X)
est = sm1.OLS(y, X).fit()

x1, x2 = np.meshgrid(np.linspace(X.LogRenda.min(), X.LogRenda.max(), 100), np.linspace(X.LogEspect.min(), X.LogEspect.max(), 100))

y_chapeu = est.params[0] + est.params[1] * x1 + est.params[2] * x2

fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-100, elev=15)

surf = ax.plot_surface(x1, x2, y_chapeu, cmap=plt.cm.RdBu_r, alpha=0.6, linewidth=0)

resid = y - est.predict(X)
ax.scatter(X[resid >= 0].LogRenda, X[resid >= 0].LogEspect, y[resid >= 0], color='black', alpha=1.0, facecolor='white')
ax.scatter(X[resid < 0].LogRenda, X[resid < 0].LogEspect, y[resid < 0], color='black', alpha=1.0)

ax.set_xlabel('Renda per Capita')
ax.set_ylabel('Espectativa de Vida')
ax.set_zlabel('IDH')
```
